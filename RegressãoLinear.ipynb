```python
import numpy as np
import matplotlib.pyplot as plt

# dados simplificados para a regressão linear
dados = np.array([
    [1.0, 1.1],
    [2.0, 3.4],
    [3.0, 5.2],
    [4.0, 6.9],
    [5.0, 9.9],
    [6.0, 11.1],
    [7.0, 13.3]
])

x = dados[:, 0].reshape(-1, 1)  # reshape para que x seja uma matriz coluna
y = dados[:, 1]

from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(x, y)

coef_angular = modelo.coef_[0]
intercepto = modelo.intercept_

x_regressao = np.linspace(min(x), max(x), 100).reshape(-1, 1)
y_regressao = modelo.predict(x_regressao)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dados Originais')
plt.plot(x_regressao, y_regressao, color='red', label=f'Linha de Regressão: y = {coef_angular:.2f}x + {intercepto:.2f}')
plt.title('Regressão Linear Simples')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
